```python
#filtering
ykappa_filter=data[data['Y-Kappa'] > 24]
ykappa_filter.head(4)
```

```python
#filtering
chiprate_filter=data[data['ChipRate']<15]
chiprate_filter.head(4)
```

```python
#filtering
blowflow_filter = data[(data['BlowFlow'] >= 1250) & (data['BlowFlow'] <= 1300)]
blowflow_filter.head(4)
```

```python
#filtering
sulphidity_filter = data[data['SulphidityL-4 '].notnull()]
sulphidity_filter.head(4)
```

```python
#missing values
missing_values = data.isnull().sum()
missing_values
```
